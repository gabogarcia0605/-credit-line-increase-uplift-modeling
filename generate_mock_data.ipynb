```python
import os
import uuid
from pathlib import Path
import numpy as np
import pandas as pd

# Crear directorio de datos si no existe
data_dir = Path("./data")
data_dir.mkdir(exist_ok=True)

print("Generando datos ficticios para el modelo de Uplift...")
np.random.seed(42)
n_samples = 14000

# Generar IDs únicos sintéticos
user_ids = [str(uuid.uuid4()) for _ in range(n_samples)]
external_ids = np.random.randint(1, 99999999, size=n_samples)
contract_numbers = [
    f"180{np.random.randint(10000, 999999)}" for _ in range(n_samples)
]

# 1. Crear usuarios_propension.csv
usuarios_df = pd.DataFrame(
    {
        "external_id": external_ids,
        "rpp_user_id": user_ids,
        "contract_number": contract_numbers,
        "treat": np.random.binomial(1, 0.2, size=n_samples),  # 20% tratamiento
        "new_credit_line": np.random.choice(
            [5000, 10000, 15000, 25000, 50000, 80000], size=n_samples
        ),
        "gender": np.random.choice(["M", "H"], size=n_samples),
        "antiguedad_meses": np.random.randint(1, 60, size=n_samples),
        "wa_enviados_30": np.random.randint(0, 15, size=n_samples),
        "wa_leidos_30": np.random.randint(0, 15, size=n_samples),
        "wa_respuestas_30": np.random.randint(0, 5, size=n_samples),
        "wa_enviados_7": np.random.randint(0, 5, size=n_samples),
        "wa_leidos_7": np.random.randint(0, 5, size=n_samples),
        "wa_respuestas_7": np.random.randint(0, 2, size=n_samples),
        "current_amount": np.random.randint(1000, 75000, size=n_samples),
        "num_ingresos_rc_90_days": np.random.randint(0, 100, size=n_samples),
        "num_ingresos_rc_30_days": np.random.randint(0, 40, size=n_samples),
        "num_ingresos_rc_7_days": np.random.randint(0, 10, size=n_samples),
        "num_ingresos_rc_d1": np.random.randint(0, 2, size=n_samples),
        "num_compras_7d": np.random.randint(0, 15, size=n_samples),
        "num_compras_30d": np.random.randint(0, 50, size=n_samples),
        "num_compras_90d": np.random.randint(0, 150, size=n_samples),
        "pctj_util_7d": np.random.uniform(0, 1, size=n_samples),
        "pctj_util_30d": np.random.uniform(0, 1, size=n_samples),
        "pctj_util_90d": np.random.uniform(0, 1, size=n_samples),
        "avg_ticket_7d": np.random.uniform(100, 1500, size=n_samples),
        "avg_ticket_30d": np.random.uniform(100, 1500, size=n_samples),
        "avg_ticket_90d": np.random.uniform(100, 1500, size=n_samples),
        "total_comprado_7d": np.random.uniform(500, 10000, size=n_samples),
        "total_comprado_30d": np.random.uniform(1000, 30000, size=n_samples),
        "total_comprado_90d": np.random.uniform(2000, 90000, size=n_samples),
        "days_since_last_purchase_90d": np.random.randint(0, 90, size=n_samples),
        "prom_saldo_dispuesto_90d": np.random.uniform(
            1000, 50000, size=n_samples
        ),
        "prom_saldo_dispuesto_30d": np.random.uniform(
            1000, 50000, size=n_samples
        ),
        "prom_saldo_dispuesto_7d": np.random.uniform(
            1000, 50000, size=n_samples
        ),
        "last_purchase_90d": np.random.randint(0, 90, size=n_samples),
        "d_incremento": np.random.binomial(
            1, 0.05, size=n_samples
        ),  # Target de conversión (5% aprox)
    }
)

# Ajustar lógicas para que los ratios no den error o cosas incoherentes
usuarios_df["current_amount"] = np.where(
    usuarios_df["current_amount"] > usuarios_df["new_credit_line"],
    usuarios_df["new_credit_line"] - 500,
    usuarios_df["current_amount"],
)
usuarios_df["treat_label"] = np.where(
    usuarios_df["treat"] == 1, "Tratamiento", "Control"
)

# 2. Crear aux_notificaciones.csv
notificaciones_df = pd.DataFrame(
    {
        "external_id": external_ids,
        "rpp_user_id": user_ids,
        "disable_push": np.random.binomial(1, 0.1, size=n_samples),
    }
)

# 3. Crear ofertas.csv
ofertas_df = pd.DataFrame(
    {
        "rpp_user_id": user_ids,
        "num_ofertas_historicas": np.random.randint(1, 10, size=n_samples),
        "num_ofertas_aceptadas": np.random.randint(0, 5, size=n_samples),
    }
)
ofertas_df["num_ofertas_no_aceptadas"] = (
    ofertas_df["num_ofertas_historicas"] - ofertas_df["num_ofertas_aceptadas"]
)

# Guardar archivos
usuarios_df.to_csv(data_dir / "usuarios_propension.csv", index=False)
notificaciones_df.to_csv(data_dir / "aux_notificaciones.csv", index=False)
ofertas_df.to_csv(data_dir / "ofertas.csv", index=False)

print("¡Archivos guardados exitosamente en la carpeta ./data/!")
```
